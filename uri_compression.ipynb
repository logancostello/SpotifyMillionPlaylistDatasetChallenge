```python
import pandas as pd
```

```python
contents = pd.read_parquet("data/original/playlist_contents.parquet")
track_metadata = pd.read_parquet("data/original/track_metadata.parquet")
```

```python
track_uri_mapping = track_metadata[["track_uri"]].reset_index(drop=True).reset_index().rename(columns={'index': 'new_track_uri'})
track_uri_mapping.head()
```

```python
artist_uri_mapping = track_metadata[["artist_uri"]].drop_duplicates().reset_index(drop=True).reset_index().rename(columns={'index': 'new_artist_uri'})
artist_uri_mapping.head()
```

```python
album_uri_mapping = track_metadata[["album_uri"]].drop_duplicates().reset_index(drop=True).reset_index().rename(columns={'index': 'new_album_uri'})
album_uri_mapping.head()
```

```python
contents = contents.merge(track_uri_mapping, on="track_uri")
contents["track_uri"] = contents["new_track_uri"]
contents = contents.drop(columns=["new_track_uri"])
contents
```

```python
track_metadata = track_metadata.merge(track_uri_mapping, on="track_uri")
track_metadata = track_metadata.merge(artist_uri_mapping, on="artist_uri")
track_metadata = track_metadata.merge(album_uri_mapping, on="album_uri")
track_metadata["track_uri"] = track_metadata["new_track_uri"]
track_metadata["artist_uri"] = track_metadata["new_artist_uri"]
track_metadata["album_uri"] = track_metadata["new_album_uri"]
track_metadata = track_metadata.drop(columns=["new_track_uri", "new_artist_uri", "new_album_uri"])
track_metadata
```

```python
track_metadata["track_uri"] = track_metadata["track_uri"].astype("uint32")
track_metadata["artist_uri"] = track_metadata["artist_uri"].astype("uint32")
track_metadata["album_uri"] = track_metadata["album_uri"].astype("uint32")

contents["track_uri"] = contents["track_uri"].astype("uint32")
```

```python
contents.to_parquet("data/original/playlist_contents.parquet")
track_metadata.to_parquet("data/original/track_metadata.parquet")
```

```python
track_uri_mapping.to_parquet("data/uri_mappings/track_uri_mappings.parquet")
artist_uri_mapping.to_parquet("data/uri_mappings/artist_uri_mappings.parquet")
album_uri_mapping.to_parquet("data/uri_mappings/album_uri_mappings.parquet")
```
