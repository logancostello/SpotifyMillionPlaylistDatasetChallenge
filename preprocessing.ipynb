```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
md = pd.read_parquet("data/original/playlist_metadata.parquet")
model = SentenceTransformer("all-MiniLM-L6-v2")


# This model is ever so slightly better, but its much slower and takes a lot of memory
# model = SentenceTransformer("all-mpnet-base-v2")
```

```python
embeddings = model.encode(md["name"].tolist(), show_progress_bar=True)
md["title_bert_embeddings"] = list(embeddings)
```
